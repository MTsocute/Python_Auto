# Reader 对象
> 用于读取 CSV 文件中的字符，reader() 参数不可以是路径，而是一个文件对象才可以

```python
import csv
import pprint

# 使用 FILE 类型创建 reader
csvfile = open('/Users/momo/Desktop/Python_Auto/第二部分/Chapter14/Data/example.csv')
reader2 = csv.reader(csvfile)

# 展示读取到的内容
row_col = list(reader2)

pprint.pprint(row_col)
```

## 1. 使用 reader\[row]\[col] 访问特定的行和列

```python
row_col[1][1]  # 第二行第二个元素
```

## 2. For 循环读取 Reader 数据

```python
import csv

csvfile = open('/Users/momo/Desktop/Python_Auto/第二部分/Chapter14/Data/example.csv')
reader3 = csv.reader(csvfile)

# 遍历每一列数据
for row in reader3:
    print('Row #' + str(reader3.line_num) + ' ' + str(row))
```

## 3. 返回所在行 reader.line_num

```python
import csv
csvfile = open('/Users/momo/Desktop/Python_Auto/第二部分/Chapter14/Data/example.csv')
reader4 = csv.reader(csvfile)

print(f'现在所在的行数为: {reader4.line_num}')
csvfile.close()
```

## 1. 写入一行数据

```python
import csv

outfile = open('/Users/momo/Desktop/Python_Auto/第二部分/Chapter14/Data/example.csv', 'w', newline='')

# 创建 Writer 对象
writer2 = csv.writer(outfile)

# 写入一行数据
writer2.writerow(['4/12/2014 3:41', 'Blueberries', 100])  # 输入的数据是一个列表
writer2.writerow(['4/12/2014 3:41', 'Blueberries', 100])  # 输入的数据是一个列表
writer2.writerow(['4/12/2014 3:41', 'Blueberries', 100])  # 输入的数据是一个列表
writer2.writerow(['4/12/2014 3:41', 'Blueberries', 100])  # 输入的数据是一个列表

outfile.close()
```

## 2. delimiter 和 lineterminator 键字数
> 用表符代替号来分单元格，并希望有两行距

```python
import csv

outfile = open('/Users/momo/Desktop/Python_Auto/第二部分/Chapter14/Data/example.tsv', 'w', newline='')

# 指定写入对象
csvwriter = csv.writer(outfile, delimiter='\t', lineterminator='\n\n')

csvwriter.writerow(['apples', 'oranges', 'grapes', 'spam'])
csvwriter.writerow(['eggs', 'bacon', 'ham', 'spam'])
csvwriter.writerow(['spam', 'spam', 'spam', 'spam'])

outfile.close()
print('Done')
```
