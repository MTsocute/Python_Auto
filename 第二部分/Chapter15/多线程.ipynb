
# threading 模块
> Python 一般都是一个程序自上而下根据程序顺序执行，假如我们代码如下，那程序time.sleep()以后的内容就无法执行

```python
import time, datetime

print("This is the beginning")

# 程序会在这地方停止，直到 2023-12-25，但是我们仍然需要其他部分的程序运行，应该如何处理呢？
while datetime.datetime.now() < datetime.datetime(2023, 12 ,25):
    time.sleep(1)
```

> 我们可以使用 python 的 threading 模块，在程序 sleep 的时候仍然执行某个部分的代码

## 创建一个 threading 执行一个函数
> 但是在 jupyter notebook 里面 thread.Thread() 创建的线程无法使用 <br>
> 因为 jupyter 在主线程结束以后就会回收所有的输出，其他晚于线程的输出结果不会被记录

```python
import threading, time

def take_a_nap():
    print('Sleeping...')
    time.sleep(2)
    print("Wake up")

# 程序一旦进入 sleep 就会调用线程
threadObj = threading.Thread(target=take_a_nap)
# 执行线程
threadObj.start()

# 等待线程结束
threadObj.join()

# 通常：程序会在最后一行终止，但是 threading 会有两个线路，全部停止才会停止
print("This is the end of program...")
```

## 向线程的目标函数传递参数

```python
# 输出字符串并添加链接符
print('cat', 'dog', 'frog', sep=' & ')
```

我们可以知道，如果向 print() 函数中 设置 sep 关键字，会确定字符串之间的连接符是 &
那么假如我们要在线程中使用 print() 函数而且还要 **传入参数，并指定关键字** ，我们应该怎么办呢？

> target: 函数名
> args: 函数里面的参数
> kwargs: 函数使用过的关键字

```python
import threading
threadObj = \
    threading.Thread(target=print,
                     args=['Cats', 'Dogs', 'Frogs'],    # 函数中的参数
                     kwargs={'sep': ' & '}              # 函数中的关键字
                     )
threadObj.start()
```

~~~python
# 这种方法创建的 thread 是没有用的
threadObj = threading.Thread(target=print('Cats', 'Dogs', 'Frogs', sep=' & '))
~~~
