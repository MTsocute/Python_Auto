# 1. 控制键盘输入内容

## 1.1. 操作键盘输入内容

```python
import pyautogui as gui

location = gui.locateOnScreen("data/manify.png")

x,y = gui.center(location)

gui.click(x,y)

# 输入内容
gui.typewrite("magnify")

exit()
```

# 2. 键名

```python
pyautogui.typewrite(['a', 'b', 'left', 'left', 'X', 'Y'])
```

> 向 typewrite() 函数传递这些键字符串的列表, 相当于按下了这些键
> 例如，以下的调用表示按 a 键，然后是 b 键，然后是左箭头两次，最后是X和Y键
**这种按法是实际在键盘上按下，所以可以执行上下移动，回撤删除等一系列操作**

## 2.2 按下和释放键盘

```python
import pyautogui as gui
from time import sleep

# 同时 keydown 实现 command + space 的效果
gui.keyDown('command')
gui.keyDown('space')

# 注意抬起来的顺序要和执行的顺序相反
gui.keyUp('space')
gui.keyUp('command')

sleep(1)

gui.typewrite("Momo super Cool~")

exit()
```

## 2.3 快捷键还有专门的函数

```python
import pyautogui as gui

# 间断下，不然没有效果
gui.hotkey('command', 'space', interval=0.25)
```
