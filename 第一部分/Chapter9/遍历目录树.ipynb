```python
import os

# 遍历 data 下的文件结构
print('-'*32)
for folderName, subfolders, filenames in os.walk('./data'):
    print('The current folder is ' + folderName)
    for subfolder in subfolders:
        print('\tSUBFOLDER OF ' + folderName + ': ' + subfolder)
    for filename in filenames:
        print('\tFILE INSIDE ' + folderName + ': '+ filename)
    print('-'*32)
```

## 更加美观的方式

```python
import os

def print_directory_tree(root_path, indent=''):
    """
        递归打印目录树结构
    """
    if not os.path.isdir(root_path):
        return

    files = os.listdir(root_path)
    files.sort()  # 按名称排序

    for i, file_name in enumerate(files):
        file_path = os.path.join(root_path, file_name)
        is_last = (i == len(files) - 1)
        marker = '└── ' if is_last else '├── '

        print(f"{indent}{marker}{file_name}")

        if os.path.isdir(file_path):
            branch = '    ' if is_last else '│   '
            sub_indent = indent + branch
            print_directory_tree(file_path, sub_indent)

# 在此处替换为要打印目录结构的根路径
root_path = './data'

print_directory_tree(root_path)
```
